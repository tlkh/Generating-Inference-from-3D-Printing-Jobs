### Plot the data in 2D

```python
%config InlineBackend.figure_format = 'retina'
import matplotlib.pyplot as plt

plt.figure(1,figsize=(12,8))
plt.xlabel('Print Time (mins)')
plt.ylabel('Filament (m)')
plt.title('Prints (Combined)')

plt.figure(2,figsize=(12,8))
plt.xlabel('Print Time (mins)')
plt.ylabel('Filament (m)')
plt.title('Prints (Failed Focus)')

plt.figure(3,figsize=(10,8))
plt.xlabel('Print Time (mins)')
plt.ylabel('Filament (m)')
plt.title('Prints (Success Only)')

for data in matrix:
    filament = [data[0][:,8]]
    time = [data[0][:,9]]
    satisfaction = [data[:,[10]]]
    result = [data[:,[11]]]
    result_success = [[1]]
    if result[0] == result_success:
        plt.figure(1)
        plt.scatter(time, filament, c="blue", alpha=0.5,)
        plt.figure(2)
        plt.scatter(time, filament, c="blue", alpha=0.05,)
        plt.figure(3)
        plt.scatter(time, filament, c="blue", alpha=1,)
    else:
        plt.figure(1)
        plt.scatter(time, filament, c="red", alpha=0.5,)
        plt.figure(2)
        plt.scatter(time, filament, c="red", alpha=1,)
        
plt.tight_layout()
plt.show()
```
